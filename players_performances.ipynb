```python
df['kickoff_time'] = pd.to_datetime(df['kickoff_time'], format='mixed', errors='coerce')

df = df.sort_values(by=['name', 'kickoff_time'])

df['rolling_avg_influence'] = df.groupby('name')['influence'].rolling(3, min_periods=1).mean().reset_index(level=0, drop=True)

team_weekly_avg_influence = df.groupby(['season_x', 'team', 'kickoff_time'])['rolling_avg_influence'].mean().reset_index()

df['rolling_avg_ict_index'] = df.groupby('name')['ict_index'].rolling(3, min_periods=1).mean().reset_index(level=0, drop=True)
team_weekly_avg_ict_index = df.groupby(['season_x', 'team', 'kickoff_time'])['rolling_avg_ict_index'].mean().reset_index()

team_weekly_avg_influence = team_weekly_avg_influence.merge(team_weekly_avg_ict_index, on=['season_x', 'team', 'kickoff_time'], how='left')
```

```python
team_weekly_avg_influence['kickoff_time'] = pd.to_datetime(team_weekly_avg_influence['kickoff_time'], utc=True)
team_weekly_avg_influence['kickoff_time'] = team_weekly_avg_influence['kickoff_time'].dt.tz_localize(None)

df_injuries['injured_since'] = pd.to_datetime(df_injuries['injured_since'], utc=True)
df_injuries['injured_since'] = df_injuries['injured_since'].dt.tz_localize(None)

df_injuries['injured_until'] = pd.to_datetime(df_injuries['injured_until'], utc=True)
df_injuries['injured_until'] = df_injuries['injured_until'].dt.tz_localize(None)
```

```python
def count_injured_players(match_date, club):
    match_date = pd.Timestamp(match_date).to_pydatetime()

    injured_players = df_injuries[df_injuries['club'] == club]

    injured_count = injured_players[
        (injured_players['injured_since'] <= match_date) &
        ((injured_players['injured_until'].isna()) | (injured_players['injured_until'] >= match_date))
    ].shape[0]
    
    return injured_count
```

```python
team_weekly_avg_influence['injured_players'] = team_weekly_avg_influence.apply(
    lambda row: count_injured_players(row['kickoff_time'], row['team']), axis=1
)
```

```python
date_format = '%d/%m/%y'

team_weekly_avg_influence['kickoff_time'] = pd.to_datetime(
    team_weekly_avg_influence['kickoff_time'], format=date_format, dayfirst=True)
matches_df['Date'] = pd.to_datetime(matches_df['Date'], format=date_format, dayfirst=True)

agg_influence = team_weekly_avg_influence.copy()
agg_influence['Date'] = agg_influence['kickoff_time'].dt.date

agg_influence = agg_influence.groupby(['Date', 'team_id'], as_index=False).agg({
    'rolling_avg_influence': 'mean',
    'rolling_avg_ict_index': 'mean',
    'injured_players': 'sum'
})

matches_df['Date'] = matches_df['Date'].dt.date

home_df = agg_influence.copy().rename(columns={
    'team_id': 'HT_Code',
    'rolling_avg_influence': 'rolling_avg_influence_HT',
    'rolling_avg_ict_index': 'rolling_avg_ict_index_HT',
    'injured_players': 'injured_players_HT'
})
matches_with_home = pd.merge(matches_df, home_df, how='left', left_on=['Date', 'HT_Code'], right_on=['Date', 'HT_Code'])

away_df = agg_influence.copy().rename(columns={
    'team_id': 'AT_Code',
    'rolling_avg_influence': 'rolling_avg_influence_AT',
    'rolling_avg_ict_index': 'rolling_avg_ict_index_AT',
    'injured_players': 'injured_players_AT'
})
final_matches_df = pd.merge(matches_with_home, away_df, how='left', left_on=['Date', 'AT_Code'], right_on=['Date', 'AT_Code'])
```

```python
final_matches_df = final_matches_df.drop(columns=['Date'])
```

```python
final_matches_df = final_matches_df.rename(columns={
    'rolling_avg_influence_HT': 'AVG_Influence_HT',
    'rolling_avg_ict_index_HT': 'AVG_ICT_Index_HT',
    'injured_players_HT': 'Injured_Players_HT',
    'rolling_avg_influence_AT': 'AVG_Influence_AT',
    'rolling_avg_ict_index_AT': 'AVG_ICT_Index_AT',
    'injured_players_AT': 'Injured_Players_AT'
})
```

```python
final_matches_df = final_matches_df[['HT_Code', 'AT_Code','FT_Result_code', 'HT_Form', 'AT_Form',
'FT_HT_Goals', 'FT_AT_Goals', 'HT_HT_Goals', 'HT_AT_Goals', 'HT_Shots',
       'AT_Shots', 'HT_Shots_target', 'AT_Shots_target', 'HT_Fouls_Comm',
       'AT_Fouls_Comm', 'HT_Corners', 'AT_Corners', 'HT_Yellows', 'AT_Yellows',
       'HT_Reds', 'AT_Reds', 'HT_W_Odds', 'Draw_Odds', 'AT_W_Odds',
       '5GAV_FT_HT_Goals', '5GAV_FT_AT_Goals', '5GAV_HT_Shots',
       '5GAV_AT_Shots', '5GAV_HT_Shots_target', '5GAV_AT_Shots_target',
       '5GAV_HT_Fouls_Comm', '5GAV_AT_Fouls_Comm', '5GAV_HT_Corners',
       '5GAV_AT_Corners', '5GAV_HT_Yellows', '5GAV_AT_Yellows', '5GAV_HT_Reds',
       '5GAV_AT_Reds'
       , 'HT_Result_code', 'AVG_Influence_HT',
       'AVG_ICT_Index_HT', 'Injured_Players_HT', 'AVG_Influence_AT',
       'AVG_ICT_Index_AT', 'Injured_Players_AT'
]]
```

```python
final_matches_df['HT_Players_Performances'] = (0.5 * final_matches_df['AVG_Influence_HT']) + (0.5 * final_matches_df['AVG_ICT_Index_HT']) - (1 * final_matches_df['Injured_Players_HT'])

final_matches_df['AT_Players_Performances'] = (0.5 * final_matches_df['AVG_Influence_AT']) + (0.5 * final_matches_df['AVG_ICT_Index_AT']) - (1 * final_matches_df['Injured_Players_AT'])
```
